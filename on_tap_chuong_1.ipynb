```python
class SV:
    def __init__(self, ma_sv, ten_sv, lop):
        self.ma_sv = ma_sv
        self.ten_sv = ten_sv
        self.lop = lop
    def info(self):
        print(f"Mã SV: {self.ma_sv}, Tên SV: {self.ten_sv}, Lớp: {self.lop}")
class SVK18A3(SV):
    def info(self):
        print(f" Mã SV: {self.ma_sv}, Tên SV: {self.ten_sv}, Lớp: {self.lop}")
#Vi du
obj1 = SV("SV1", "Nguyen Van A", "K18A3")
obj2 = SVK18A3("SV2", "Tran Thi B", "K18A3")
obj1.info()  
obj2.info()
```
